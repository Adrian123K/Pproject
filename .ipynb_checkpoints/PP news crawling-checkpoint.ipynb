```python
def head_scr():

    import pandas as pd
    import urllib.request 
    from bs4 import BeautifulSoup 

    params1=[]
    title=''

    dt_index = pd.date_range(start='20200131', end='20200331')
    dueday = dt_index.strftime("%Y%m%d").tolist()
    
    h=open('headline1Q.txt','w',encoding='utf-8')
        
    for day in dueday :
        list_url=f"https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date={day}"
        url=urllib.request.Request(list_url) 
        rs=urllib.request.urlopen(url).read().decode("cp949") 
        soup=BeautifulSoup(rs, "html.parser")

        rs1=soup.find_all('a', class_='nclicks(rnk.sci)')

        for j in rs1:
            title += j.get_text('title',strip=True)+'\n'            
            params1.append(j.get_text('title',strip=True))
                        
        head = [i for i in params1 if i]
        
    h.write(title)
    h.close()

    return head
head_scr()
```
